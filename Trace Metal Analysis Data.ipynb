# Sample Prep

```python

temp_df=[]
#for item in ['1', '2', '3', '4 & 5', '6', '7, 8']:
for item in ['1', '6', '7, 8']:
    print(item)
    sam=pd.read_excel('Cations_list.xlsx', sheet_name= item)

    sam=sam.rename(columns={'Stations':'Sample'})
    sam['Sample']=sam['Sample'].ffill()
    sam.columns = sam.columns.to_series().mask(lambda x: x.str.startswith('Unnamed')).ffill()
    sam=sam.iloc[1:, 1:]
 
    temp_df.append(sam)
```

```python
sam
```

```python
sample_preparation=pd.concat(temp_df).reset_index(drop=True)
sample_preparation
```

```python
sample_preparation.head()
```

```python
sample_preparation.columns
```

```python
for col in ['1st Triplicate Initial ', '1st Triplicate final',
       '2nd Triplicate initial', '2nd Triplicate final',
       '3rd Triplicate initial', '3rd Triplicate final',]:
    sample_preparation[col]=sample_preparation[col].astype(str).str.replace(',', '')
```

```python
#sample_preparation=sample_preparation.drop(axis=0,index=0)
```

```python
sample_preparation['1st triplicate dil factor']=sample_preparation['1st Triplicate final'].astype(float)/sample_preparation['1st Triplicate Initial '].astype(float)
sample_preparation
```

```python
sample_preparation['3rd Triplicate initial']=sample_preparation['3rd Triplicate initial'].replace(0,np.nan)
```

```python
sample_preparation['2nd triplicate dil factor']=sample_preparation['2nd Triplicate final'].astype(float)/sample_preparation['2nd Triplicate initial'].astype(float)
sample_preparation['3rd triplicate dil factor']=sample_preparation['3rd Triplicate final'].astype(float)/sample_preparation['3rd Triplicate initial'].astype(float)
sample_preparation
```

```python
sample_preparation.columns
```

```python
sample_preparation =sample_preparation.drop(columns=[ 'Cruise Type','1st Triplicate Initial ', '1st Triplicate final',
       '2nd Triplicate initial', '2nd Triplicate final',
       '3rd Triplicate initial', '3rd Triplicate final'])
sample_preparation
```
